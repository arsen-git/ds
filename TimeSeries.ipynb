```python
import pandas as pd

import pandas_datareader as dr

df = dr.get_data_yahoo('AAPl', start='2015-01-01', end='2018-12-30')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kapanis = df['Close']
kapanis.head()
```

```python
df.shape
```

```python
kapanis.plot()
```

```python
kapanis.index = pd.DatetimeIndex(kapanis.index)
kapanis.plot()
```

```python
yedi = kapanis.rolling(window=7).mean()
on = kapanis.rolling(window=10).mean()
yirmi = kapanis.rolling(window=20).mean()
otuz = kapanis.rolling(window=30).mean()
```

```python
yedi.head(10)
```

```python
on.head(15)
```

```python
kapanis.plot(style='-')
yedi.plot(style='--')
on.plot(style='-.')
yirmi.plot(style=':')
otuz.plot()
plt.legend(['input', 'yedi', 'on', 'yirmi', 'otuz'], loc='upper left')
plt.xlim('2017', '2018')
```
