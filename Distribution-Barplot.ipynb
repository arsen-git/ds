```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe().T
```

```python
df.cut = df.cut.astype(pd.CategoricalDtype(ordered=True))
df.color = df.color.astype(pd.CategoricalDtype(ordered=True))
df.clarity = df.clarity.astype(pd.CategoricalDtype(ordered=True))
df.dtypes
```

```python
print(df.cut.head(1))
print(df.color.head(1))
print(df.clarity.head(1))
```

```python
cut_kategoriler = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
```

```python
df.cut = df.cut.astype(pd.CategoricalDtype(categories=cut_kategoriler, ordered=True))
```

```python
df.cut.head(1)
```

# Barplot

```python
(df['cut']
    .value_counts()
    .plot.barh()
    .set_title('Barplot Kategorik Değişken Göstermek İçin Kullanılır'));
```

```python
(df['clarity']
    .value_counts()
    .plot.barh()
    .set_title('Barplot Kategorik Değişken Göstermek İçin Kullanılır'));
```

```python
sns.catplot(x='cut', y='price', hue='color', data=df)
```
